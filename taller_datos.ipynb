```python
indicador = {"NY.GDP.PCAP.CD": "PIB per cápita"}
fecha_inicio = datetime.datetime(2010, 1, 1)
fecha_fin = datetime.datetime(2023, 1, 1)
Paises = ["COL", "ECU", "BRA", "CHL", "MEX", "PER"]
dfPIB = wbdata.get_dataframe(indicador, country = Paises, date=(fecha_inicio, fecha_fin), skip_cache=True).reset_index()

print(dfPIB)
print(dfPIB.head())
print(dfPIB.tail(10))
print(dfPIB.sample(3))
print(dfPIB.iloc[1:5,1])
print(len(dfPIB))
print(dfPIB.shape)
print(dfPIB.columns)
print(dfPIB.dtypes)
print(dfPIB.info())
```
